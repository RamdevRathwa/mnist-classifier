```python
# MNIST Handwritten Digit Classification Project

# 1. Data Collection & Preprocessing

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize pixel values (0-255 to 0-1)
x_train, x_test = x_train / 255.0, x_test / 255.0

# Convert labels to categorical (one-hot encoding)
y_train, y_test = to_categorical(y_train, 10), to_categorical(y_test, 10)

# Split training data into training (80%) and validation (20%)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)

print(f"Training Data Shape: {x_train.shape}, Validation Data Shape: {x_val.shape}, Test Data Shape: {x_test.shape}")

# 2. Exploratory Data Analysis (EDA)

def visualize_samples(x, y, n_samples=10):
    plt.figure(figsize=(10, 5))
    for i in range(n_samples):
        plt.subplot(2, 5, i + 1)
        plt.imshow(x[i], cmap='gray')
        plt.title(f"Label: {np.argmax(y[i])}")
        plt.axis('off')
    plt.show()

visualize_samples(x_train, y_train)

# Check class distribution
unique, counts = np.unique(np.argmax(y_train, axis=1), return_counts=True)
print("Class Distribution:", dict(zip(unique, counts)))

# Flatten images for traditional ML models
x_train_flat = x_train.reshape(x_train.shape[0], -1)
x_val_flat = x_val.reshape(x_val.shape[0], -1)
x_test_flat = x_test.reshape(x_test.shape[0], -1)

# Convert labels back to integers
y_train_labels = np.argmax(y_train, axis=1)
y_val_labels = np.argmax(y_val, axis=1)
y_test_labels = np.argmax(y_test, axis=1)

# 3. Logistic Regression Model
log_model = LogisticRegression(max_iter=1000)
log_model.fit(x_train_flat, y_train_labels)

# Evaluate on test set
y_test_pred_log = log_model.predict(x_test_flat)
print("Test Accuracy (Logistic Regression):", accuracy_score(y_test_labels, y_test_pred_log))
print(confusion_matrix(y_test_labels, y_test_pred_log))

# 4. Support Vector Machine (SVM) Model
svm_model = SVC(kernel='linear')
svm_model.fit(x_train_flat, y_train_labels)

y_test_pred_svm = svm_model.predict(x_test_flat)
print("Test Accuracy (SVM):", accuracy_score(y_test_labels, y_test_pred_svm))
print(confusion_matrix(y_test_labels, y_test_pred_svm))

# 5. Random Forest Model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train_flat, y_train_labels)

y_test_pred_rf = rf_model.predict(x_test_flat)
print("Test Accuracy (Random Forest):", accuracy_score(y_test_labels, y_test_pred_rf))
print(confusion_matrix(y_test_labels, y_test_pred_rf))

# 6. K-Nearest Neighbors (KNN) Model
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(x_train_flat, y_train_labels)

y_test_pred_knn = knn_model.predict(x_test_flat)
print("Test Accuracy (KNN):", accuracy_score(y_test_labels, y_test_pred_knn))
print(confusion_matrix(y_test_labels, y_test_pred_knn))

# 7. Convolutional Neural Network (CNN) Model

x_train_cnn = x_train.reshape(-1, 28, 28, 1)
x_val_cnn = x_val.reshape(-1, 28, 28, 1)
x_test_cnn = x_test.reshape(-1, 28, 28, 1)

data_gen = ImageDataGenerator(rotation_range=10, zoom_range=0.1, width_shift_range=0.1, height_shift_range=0.1)
data_gen.fit(x_train_cnn)

cnn_model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='softmax')
])

cnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

cnn_history = cnn_model.fit(data_gen.flow(x_train_cnn, y_train, batch_size=64), epochs=15, validation_data=(x_val_cnn, y_val))

cnn_model.save("mnist_cnn_model.h5")
print("Model saved as mnist_cnn_model.h5")

loaded_model = load_model("mnist_cnn_model.h5")
print("Model loaded successfully")

cnn_test_loss, cnn_test_acc = loaded_model.evaluate(x_test_cnn, y_test)
print("Test Accuracy (CNN):", cnn_test_acc)

y_pred_cnn = np.argmax(loaded_model.predict(x_test_cnn), axis=1)
print(classification_report(y_test_labels, y_pred_cnn))
print(confusion_matrix(y_test_labels, y_pred_cnn))
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
import tkinter as tk
from tkinter import Canvas, Button, Label, Frame
from PIL import Image, ImageGrab

# Load the trained model
model = load_model("mnist_cnn_model.h5")
print("Model loaded successfully!")

# Digit Classifier GUI
class DigitClassifierApp:
    def __init__(self, root, model):
        self.root = root
        self.model = model

        # Main Frame
        main_frame = Frame(root, padx=20, pady=20)
        main_frame.grid(row=0, column=0)

        # Canvas for drawing
        self.canvas = Canvas(main_frame, width=280, height=280, bg='black', highlightthickness=2, relief="ridge")
        self.canvas.grid(row=0, column=0, columnspan=2, pady=10)

        # Prediction Label
        self.prediction_label = Label(main_frame, text="Draw a digit", font=("Helvetica", 24), fg="blue")
        self.prediction_label.grid(row=1, column=0, columnspan=2, pady=20)

        # Buttons (Predict and Clear)
        self.button_predict = Button(main_frame, text="Predict", command=self.predict_digit, font=("Arial", 14), bg="#4CAF50", fg="white", padx=10)
        self.button_predict.grid(row=2, column=0, padx=10, pady=10, sticky="ew")

        self.button_clear = Button(main_frame, text="Clear", command=self.clear_canvas, font=("Arial", 14), bg="#F44336", fg="white", padx=10)
        self.button_clear.grid(row=2, column=1, padx=10, pady=10, sticky="ew")

        # Bind mouse drag event
        self.canvas.bind("<B1-Motion>", self.draw)

    def draw(self, event):
        x, y = event.x, event.y
        self.canvas.create_oval(x, y, x + 8, y + 8, fill='white', outline='white')

    def clear_canvas(self):
        self.canvas.delete("all")
        self.prediction_label.config(text="Draw a digit")

    def predict_digit(self):
        # Capture canvas content
        x = self.root.winfo_rootx() + self.canvas.winfo_x()
        y = self.root.winfo_rooty() + self.canvas.winfo_y()
        x1 = x + self.canvas.winfo_width()
        y1 = y + self.canvas.winfo_height()

        img = ImageGrab.grab(bbox=(x, y, x1, y1)).convert("L")
        img = img.resize((28, 28))
        img = np.array(img) / 255.0
        img = img.reshape(1, 28, 28, 1)

        # Make prediction and get confidence
        predictions = self.model.predict(img)
        predicted_digit = np.argmax(predictions)
        confidence = np.max(predictions) * 100

        # Display prediction with confidence
        self.prediction_label.config(text=f"Predicted: {predicted_digit} (Confidence: {confidence:.2f}%)")

# Run the GUI
root = tk.Tk()
root.title("MNIST Digit Classifier")
root.resizable(False, False)  # Disable resizing for better UI stability

app = DigitClassifierApp(root, model)
root.mainloop()
```
